```python
from PIL import Image as im
import matplotlib.pyplot as plt
```

```python
image_path="/content/drive/MyDrive/ACV/Zoro.jpg"
```

```python
def adjust_brightness(image, factor):
    # Open the image
    img = im.open(image)

    # Get the dimensions of the image
    width, height = img.size

    # Access pixel data
    pixels = img.load()

    # Apply brightness adjustment
    for x in range(width):
        for y in range(height):
            # Get pixel value
            r, g, b = pixels[x, y]

            # Adjust brightness
            r = min(255, max(0, int(r * factor)))
            g = min(255, max(0, int(g * factor)))
            b = min(255, max(0, int(b * factor)))

            # Update pixel value
            pixels[x, y] = (r, g, b)

    # Show the adjusted image
    plt.imshow(img)
    plt.axis('off')  # Hide axes
    plt.show()
```

```python
def adjust_contrast(image, factor):
    # Open the image
    img = im.open(image)

    # Get the dimensions of the image
    width, height = img.size

    # Access pixel data
    pixels = img.load()

    # Apply contrast adjustment
    for x in range(width):
        for y in range(height):
            # Get pixel value
            r, g, b = pixels[x, y]

            # Adjust contrast
            r = min(255, max(0, int((r - 128) * factor + 128)))
            g = min(255, max(0, int((g - 128) * factor + 128)))
            b = min(255, max(0, int((b - 128) * factor + 128)))

            # Update pixel value
            pixels[x, y] = (r, g, b)

    # Show the adjusted image
    plt.imshow(img)
    plt.axis('off')  # Hide axes
    plt.show()
```

```python
def convert_to_grayscale(image):
    # Open the image
    img = im.open(image)

    # Get the dimensions of the image
    width, height = img.size

    # Access pixel data
    pixels = img.load()

    # Convert the image to grayscale
    for x in range(width):
        for y in range(height):
            # Get pixel value
            r, g, b = pixels[x, y]

            # Calculate average value
            gray = int((r + g + b) / 3)

            # Update pixel value
            pixels[x, y] = (gray, gray, gray)
    # Show the grayscale image
    plt.imshow(img)
    plt.axis('off')  # Hide axes
    plt.show()
```

```python
adjust_brightness(image_path, 1.5)
```

```python
adjust_contrast(image_path, 1.5)
```

```python
convert_to_grayscale(image_path)
```
